```python
##Task 1 ##

#1. Variables and Advanced Operations:

#Create variables of different types (int, float, complex, string, list, tuple, set, dictionary).
int_var=10
float_var=10.0
com_var=3 + 2j
str_var='Amani Shihap'
my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)
my_set = {1, 2, 3, 4, 5}
my_dict = {"key1": "value1", "key2": "value2"}

#perform some arithmetic operations on numeric variables, including complex numbers
# Basic arithmetic operations
a = 29
b = 10
sum_ab = a + b
sub_ab = a - b
multi_ab = a * b
division_ab = a / b
print(f'sum is: {sum_ab}, sub is: {sub_ab}, multiplication is: {multi_ab}, and division is: {division_ab}\n')

# Operations with complex numbers
c1 = 2 + 3j
c2 = 4 + 5j
sum_c = c1 + c2
sub_c = c2 - c1
multi_c = c1 * c2
division_c = c2 / c1
print(f'Complex sum is: {sum_c}, complex sub is: {sub_c}, complex multiplication is: {multi_c}, and complex division is: {division_c}\n')

#advanced operators
x = 22
print('Raw x is:',x)
x += 1 
print('After adding 1 x is:',x)
x -= 15
print('After sub 15 x is:',x)
x *= 2
print('Dublicated x is:',x)
x /= 4 
print('x divided by 4 is:',x)

#logical operators
a = True
b = False
and_operation = a and b
or_operation = a or b
not_operation = not a
print(f'And:{and_operation}, OR:{or_operation}, NOT:{not_operation}\n')

#Operator Precedence
result = 10 + 2 * 3  
print('Result of "10 + 2 * 3" is:',result)
result = (10 + 2) * 3  # Parentheses change the precedence, result is 36
print('Result of "(10 + 2 )* 3" is:',result)

#2.Advanced condition
#Program to Determine Leap Year
def is_leap_year(year):
    """Determine if a given year is a leap year."""
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Example usage:
year = int(input("Enter a year to check if it is a leap year: "))
if is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")

# Function to Categorize a List of Mixed Types
def categorize_mixed_list(mixed_list):
    int_list = []
    float_list = []
    string_list = []

    for element in mixed_list:
        try:
            if isinstance(element, int):
                int_list.append(element)
            elif isinstance(element, float):
                float_list.append(element)
            elif isinstance(element, str):
                string_list.append(element)
            else:
                print(f"Unrecognized type for element: {element}")
        except Exception as e:
            print(f"An error occurred: {e} for element: {element}")

    return int_list, float_list, string_list
print("Welcome to the Mixed List Categorizer!")
print("You can enter a list of mixed types, and I will categorize them into integers, floats, and strings.")
user_input = input("Enter your list (e.g., [1, 2.5, 'hello', 3]): ")
try:
    mixed_list = eval(user_input)
    if not isinstance(mixed_list, list):
        raise ValueError("The input is not a valid list.")
except Exception as e:
    print(f"Invalid input: {e}")
else:
    integers, floats, strings = categorize_mixed_list(mixed_list)
    print("\nHere are the categorized lists:")
    print("Integers:", integers)
    print("Floats:", floats)
    print("Strings:", strings)

#3. Nested Loops and Control Statements:
#generate Pascal's Triangle up to 10 levels
def generate_pascals_triangle(levels):
    triangle = []
    for i in range(levels):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = triangle[i - 1][j - 1] + triangle[i - 1][j]
        triangle.append(row)

    return triangle
levels = 10
triangle = generate_pascals_triangle(levels)

print("Pascal's Triangle (up to 10 levels):")
for row in triangle:
    print(row)

# calculate the factorial of a number
def calculate_factorial(n):
    if n < 0:
        raise ValueError("Factorial is not defined for negative numbers.")
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial
try:
    user_input = input("Enter a non-negative integer to calculate its factorial: ")
    number = int(user_input)
    result = calculate_factorial(number)
    print(f"The factorial of {number} is {result}")
except ValueError as e:
    print(f"Invalid input: {e}")

#print prime numbers within a given range
def is_prime(num):
    """Check if a number is a prime number."""
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def print_primes_in_range(start, end):
    print(f"Prime numbers between {start} and {end} (excluding multiples of 3 or 5):")
    for num in range(start, end + 1):
        if num % 3 == 0 or num % 5 == 0:
            continue
        if is_prime(num):
            print(num, end=' ')
    print()
start = 10
end = 50
print_primes_in_range(start, end)
```

```python
##Task 2##

#1. Complex Classes and Objects:
#Create a class BankAccount with attributes for account number, balance
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}.")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > self.balance:
            raise ValueError("Insufficient funds for withdrawal.")
        self.balance -= amount
        print(f"Withdrew {amount}. New balance is {self.balance}.")

    def check_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Number: {self.account_number}, Account Holder: {self.account_holder}, Balance: {self.balance}"
try:
    account1 = BankAccount("123456", "Alice", 1000.0)
    print(account1)
    account1.deposit(500.0)
    account1.withdraw(200.0)
    try:
        account1.withdraw(1500.0)
    except ValueError as e:
        print(f"Error: {e}")
    print(f"Current balance: {account1.check_balance()}")
    account2 = BankAccount("789012", "Bob")
    print(account2)
    account2.deposit(300.0)
    try:
        account2.withdraw(-50.0)  # Invalid operation
    except ValueError as e:
        print(f"Error: {e}")
    print(f"Current balance for {account2.account_holder}: {account2.check_balance()}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

#2. Inheritance and Multiple Inheritance:
# Base class: BankAccount
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}.")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > self.balance:
            raise ValueError("Insufficient funds for withdrawal.")
        self.balance -= amount
        print(f"Withdrew {amount}. New balance is {self.balance}.")

    def check_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Number: {self.account_number}, Account Holder: {self.account_holder}, Balance: {self.balance}"
class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, interest_rate=0.0):
        super().__init__(account_number, account_holder, balance)
        self.interest_rate = interest_rate
    def calculate_interest(self, periods):
        interest = self.balance * (self.interest_rate / 100) * periods
        return interest
    def __str__(self):
class LoanAccount:
    def __init__(self, loan_amount, interest_rate):
        self.loan_amount = loan_amount
        self.interest_rate = interest_rate

    def calculate_loan_interest(self, periods):
        interest = self.loan_amount * (self.interest_rate / 100) * periods
        return interest
    def __str__(self):
        return f"Loan Amount: {self.loan_amount}, Interest Rate: {self.interest_rate}%"
class ComprehensiveAccount(SavingsAccount, LoanAccount):
    def __init__(self, account_number, account_holder, balance=0.0, savings_interest_rate=0.0, loan_amount=0.0, loan_interest_rate=0.0):
        SavingsAccount.__init__(self, account_number, account_holder, balance, savings_interest_rate)
        LoanAccount.__init__(self, loan_amount, loan_interest_rate)
    def __str__(self):
        return f"{SavingsAccount.__str__(self)}, {LoanAccount.__str__(self)}"
try:
    savings_account = SavingsAccount("1001", "Charlie", 5000.0, 3.5)
    print(savings_account)
    interest = savings_account.calculate_interest(2)
    print(f"Calculated interest for 2 periods: {interest}")
    comprehensive_account = ComprehensiveAccount("2001", "Diana", 10000.0, 4.0, 5000.0, 6.5)
    print(comprehensive_account)
    savings_interest = comprehensive_account.calculate_interest(1)
    print(f"Calculated savings interest for 1 period: {savings_interest}")
    loan_interest = comprehensive_account.calculate_loan_interest(1)
    print(f"Calculated loan interest for 1 period: {loan_interest}")

except Exception as e:
    print(f"An error occurred: {e}")
    
#3. Polymorphism and Method Overriding:
# Base class: Account
class Account:
    def __init__(self, account_number, account_holder, balance=0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}.")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > self.balance:
            raise ValueError("Insufficient funds for withdrawal.")
        self.balance -= amount
        print(f"Withdrew {amount}. New balance is {self.balance}.")

    def check_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Number: {self.account_number}, Account Holder: {self.account_holder}, Balance: {self.balance}"
class SavingsAccount(Account):
    def __init__(self, account_number, account_holder, balance=0.0, interest_rate=0.0):
        super().__init__(account_number, account_holder, balance)
        self.interest_rate = interest_rate

    def calculate_interest(self, periods):
        interest = self.balance * (self.interest_rate / 100) * periods
        return interest

    def __str__(self):
        return f"{super().__str__()}, Interest Rate: {self.interest_rate}%"

class LoanAccount(Account):
    def __init__(self, account_number, account_holder, balance=0.0, loan_limit=0.0):
        super().__init__(account_number, account_holder, balance)
        self.loan_limit = loan_limit

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if (self.balance - amount) < -self.loan_limit:
            raise ValueError(f"Cannot overdraw beyond the loan limit of {self.loan_limit}.")
        self.balance -= amount
        print(f"Withdrew {amount}. New balance is {self.balance}.")

    def __str__(self):
        return f"{super().__str__()}, Loan Limit: {self.loan_limit}"

def perform_transactions(account):
    print("\nPerforming transactions on account:")
    print(account)

    try:
        account.deposit(500)
        account.withdraw(300)
        account.withdraw(1000) 

    except ValueError as e:
        print(f"Transaction error: {e}")

    print(f"Final balance: {account.check_balance()}")
savings_account = SavingsAccount("3001", "Eve", 1000, 2.0)
loan_account = LoanAccount("4001", "Frank", 500, 300)
perform_transactions(savings_account)
perform_transactions(loan_account)

#4. Advanced File Handling and Data Persistence:
import json
from cryptography.fernet import Fernet
import os

def generate_key():
    key = Fernet.generate_key()
    with open("secret.key", "wb") as key_file:
        key_file.write(key)

def load_key():
    return open("secret.key", "rb").read()

def encrypt_data(data, key):
    fernet = Fernet(key)
    encrypted_data = fernet.encrypt(data.encode())
    return encrypted_data

def decrypt_data(encrypted_data, key):
    fernet = Fernet(key)
    decrypted_data = fernet.decrypt(encrypted_data).decode()
    return decrypted_data

def save_account_data(account, filename, key):
    data = {
        "account_number": account.account_number,
        "account_holder": account.account_holder,
        "balance": account.balance
    }
    json_data = json.dumps(data)
    encrypted_data = encrypt_data(json_data, key)
    with open(filename, "wb") as file:
        file.write(encrypted_data)

# Load bank account data from a JSON file with decryption
def load_account_data(filename, key):
    with open(filename, "rb") as file:
        encrypted_data = file.read()
    json_data = decrypt_data(encrypted_data, key)
    data = json.loads(json_data)
    return data

# BankAccount class
class BankAccount:
    def __init__(self, account_number, account_holder, balance=0.0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > self.balance:
            raise ValueError("Insufficient funds for withdrawal.")
        self.balance -= amount

    def check_balance(self):
        return self.balance

    def __str__(self):
        return f"Account Number: {self.account_number}, Account Holder: {self.account_holder}, Balance: {self.balance}"

def main():
    key = load_key()
    account = BankAccount("123456", "Alice", 1000.0)
    print("Original Account Data:")
    print(account)
    save_account_data(account, "account_data.json", key)
    loaded_data = load_account_data("account_data.json", key)
    print("\nLoaded Account Data:")
    print(loaded_data)
    loaded_account = BankAccount(
        loaded_data["account_number"],
        loaded_data["account_holder"],
        loaded_data["balance"]
    )
    print("\nBankAccount Object from Loaded Data:")
    print(loaded_account)

if __name__ == "__main__":
    main()
```
